```python
test_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv")
train_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv")
```

```python
y = train_data.SalePrice
```

```python
train_data_new = train_data.drop(['Id'],axis=1)
```

```python
features = [ 'LotArea','YearBuilt','1stFlrSF','2ndFlrSF',
             'TotRmsAbvGrd','Fireplaces','GarageCars',
             'GarageArea','PoolArea']
```

```python
X=train_data_new[features]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
house_model = RandomForestRegressor(random_state  = 1)
house_model.fit(X,y)
```

```python
test_X=test_data[features]
```

```python
test_X['GarageCars'].fillna(1.77 , inplace=True)
```

```python
test_X['GarageArea'].fillna(472.77 , inplace = True)
```

```python
predictions = house_model.predict(test_X)
```

```python
output = pd.DataFrame({'Id':test_data.Id , 'SalePrice':predictions})
output.to_csv('house_prices_submission.csv', index=False)
print("Your submission was successfully saved!")
```
